```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_house_sales.csv')
df.head()
```

## Geographic Analysis

```python
# Calculate average sale price by city
avg_price_by_city = df.groupby('city')['sale_price'].mean().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_price_by_city.values, y=avg_price_by_city.index, legend=False, color='blue')
plt.title('Average Sale Price by City')
plt.xlabel('Average Sale Price ($)')
plt.ylabel('City')
plt.show()
```

```python
# Explore the correlation between the city and the area
plt.figure(figsize=(10, 6))
sns.boxplot(x='city', y='area', data=df, color='skyblue')
plt.title('Area (sq. ft.)')
plt.xlabel('City')
plt.ylabel('Sale Price ($)')
plt.xticks(rotation=90)
plt.show()
```

```python
# Explore the correlation between the city and the number of rooms
plt.figure(figsize=(10, 6))
sns.boxplot(x='city', y='bedrooms', data=df, color='skyblue')
plt.title('Number of Rooms')
plt.xlabel('City')
plt.ylabel('Sale Price ($)')
plt.xticks(rotation=90)
plt.show()
```
